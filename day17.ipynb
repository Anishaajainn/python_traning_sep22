```python
import matplotlib.pyplot as pit
import numpy as np
```

```python
#types of graphs
#1. Plot -   line graph, between two parameters
#2. bar -    bar graph of double parameter 
#3. hist -   histogram of a bunch of data i.e. frequency of double param
#4. scatter - dot graph between two params 
#5. boxplot/box - statstical overview in form of box for a single params
#                mean, outliers , qunatiles(25%,75%), IQR
#                (Optional - categorical data)
#6. stem - more like bar graph, lines instead of bars
#7. violinplot - violin shaped statistical overview graph spread of data 
#                (more like hist)
#                 single parameter (optional cat-data)         
```

```python
x = np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y = x**3
```

```python
y2=x**4
```

```python
pit.plot(x,y)
#pit.show()
pit.title('plot')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.bar(x,y)
pit.title('bar Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.hist(x,y)
pit.title('hist Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.scatter(x,y)
pit.title('scatter Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.boxplot(x)
pit.title('barplot Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.stem(x,y)
pit.title('stem Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
pit.violinplot(x)
pit.title('violinplot Graph')
pit.grid()
pit.xlabel('x--->')
pit.xlabel('y--->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
y2=x**4
```

```python
pit.plot(x,y)
pit.show()
pit.plot(x,y2)
pit.show()
```

```python
pit.plot(x,y,x,y2)
pit.legend(['y=x**3','y2=x**4'])
pit.show()
```

```python
fig = pit.figure(figsize=(10,5))
pit.plot(x,y,x,lw=3,marker='o',markersize=15,markerfacecolor='yellow')
pit.plot(x,y,marker='v',markersize=10)
pit.legend(['y=x**3','y2=x**4'])
pit.show()
```
